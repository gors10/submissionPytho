### Explore ...

```python
orders_items_df = pd.merge(
    left= order_items,
    right= products_df,
    how="left",
    left_on="product_id",
    right_on= "product_id"
)

orders_items_df.head()
```

```python
data_1 = pd.merge(
    left= order_df,
    right= orders_items_df,
    how="left",
    left_on= "order_id",
    right_on= "order_id"
)

data_1.head()
```

```python
data_1.order_status.head(20)
```

```python
status_counts = data_1["order_status"].unique()
print(status_counts)
```

```python
approved_orders = data_1[orders_df['order_status'] == 'approved']
```

```python
result = approved_orders.groupby("product_category_name").agg(
    min_price=("price", "min"),
    max_price=("price", "max")
)

print(result)
```

```python
data_2 = pd.merge(
    left= order_df,
    right= customers_df,
    how="left",
    left_on= "customer_id",
    right_on= "customer_id"
)

data_2.head()
```

```python
status_counts = data_2["customer_state"].unique()
print(status_counts)
```

```python
result2 = data_2.groupby("customer_state").agg({
    "order_id": "nunique",
    "order_approved_at": "count"  # Menggunakan count untuk menghitung jumlah transaksi
})
print(result2)
```
